<h2>Ejercicio 2 <h2>

```python
'''
Fecha (un registro por día)
Cantidad_articulos : cantidad de artículos distintos 
para cada fecha
Total_importe: suma de los importes para cada fecha
'''
importe_x_fecha = df.groupby('fecha').sum()['importe'].reset_index()
arts_x_fecha = df.groupby(['fecha']).nunique()['item'].reset_index()

df_importe = pd.DataFrame(importe_x_fecha)
df_arts = pd.DataFrame(arts_x_fecha)
new_df = pd.merge( df_arts, df_importe, how="outer") 

new_df.columns = ['fecha', 'cantidad_articulos', 'total_importe']
new_df
```

<h2>Ejercicio 3<h2>

```python
'''
Si la variable grupo es igual a 15 LIMPIEZA DEL HOGAR o 18 PERFUMERIA, 
entonces SECTOR = “NO COMESTIBLES”

Si la variable grupo es igual a 4 LACTEOS o 9 FIAMBRERIA, 
entonces SECTOR = “PERECEDEROS”

El resto de los casos SECTOR = grupo
'''
def rename(grupo):
    if(grupo == "15 LIMPIEZA DEL HOGAR" or grupo == "18 PERFUMERIA"):
        return "NO COMESTIBLES"
    elif(grupo == "4 LACTEOS" or "9 FIAMBRERIA"):
        return "PERECEDEROS"
    else:
        return "grupo"
    

df["SECTOR"] = [rename(a) for a in df["grupo"]]
df
```

<h2>Ejercicio 4<h2>

```python
'''
Debe haber dos columnas para cada uno de los sectores creados en el ejercicio 3, dichas columna serán:
- promedio de precio
- suma de importe
Los nombres de las columnas son discrecionales y a decisión de Uds.


Las filas deben ser los meses obtenidos de la columna fecha
'''

df2 = df.copy()

# Formateo de fecha

df2['fecha'] = pd.to_datetime(df['fecha'])
fechas = df2['fecha'].dt.strftime('%b-%Y')
df2['fecha'] = fechas

new_df2 = df2.groupby(['fecha','SECTOR']).agg(['sum', 'mean']).reset_index()
new_df2[['fecha','SECTOR','item','precio']].groupby('SECTOR')
new_df2
```
